```python
import pandas as pd
import random

# Sample data for the DataFrame
data = {
    'ID': range(1, 21),
    'first_name': [random.choice(['John', 'Jane', 'Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Hannah']) for _ in range(20)],
    'last_name': [random.choice(['Smith', 'Doe', 'Brown', 'Taylor', 'Johnson', 'White', 'Clark', 'Lewis', 'Walker', 'Hall']) for _ in range(20)],
    'salary': [random.randint(50000, 150000) for _ in range(20)],
    'department_id': [random.randint(1, 10) for _ in range(20)]
}

# Creating the DataFrame
df = pd.DataFrame(data)
df.head(10)  # Displaying the first 10 rows
```

```python
sort_salaries = df.sort_values(['salary'], ascending=False)
```

```python
sort_salaries
```

```python
current_salaries = sort_salaries.drop_duplicates(['ID'])
```

```python
result = current_salaries.sort_values(['ID'])
result
```

**Approach 2**

```python
latest_salary_df = df.groupby('ID').apply(lambda x: x.loc[x['salary'].idxmax()]).reset_index(drop=True)
latest_salary_df
```

```python
# Step 2: Sort by ID
latest_salary_df = latest_salary_df.sort_values('ID')
latest_salary_df
```

```python
import re
```

```python
# Creating a sample DataFrame with 'filename' and 'contents'
data = {
    'filename': ['file1.txt', 'file2.txt', 'file3.txt'],
    'contents': [
        "The bull charged through the market. The bear waited.",
        "The market was quiet with no bull or bear in sight. This is bullshit",
        "Bull and bear are often used to describe markets."
    ]
}

# Create the DataFrame
df_google_file_store = pd.DataFrame(data)

# Display the DataFrame
df_google_file_store.head()
```

```python
df_google_file_store['bull'] = df_google_file_store['contents'].apply(lambda text: len(re.findall(r'\bbull\b', text)))
df_google_file_store['bear'] = df_google_file_store['contents'].apply(lambda text: len(re.findall(r'\bbear\b', text)))
```

```python
df_google_file_store
```

```python
df_google_file_store[['bull','bear']].sum().reset_index()
```

### Solution 2:

```python
df_google_file_store.assign(bear = lambda x : x.contents.str.count('bear')).assign(bull = lambda x: x.contents.str.count('bull'))[['bull','bear']].sum().reset_index()
```

### Solution 3:

```python
bull_count = df_google_file_store['contents'].str.count('bull').sum()
bear_count = df_google_file_store['contents'].str.count('bear').sum()
result = {'index':['bull', 'bear'], '0': [bull_count, bear_count]}
print(pd.DataFrame(result))
```

```python
# Sample dataset
data = {
    'id': [1, 2, 3, 4, 5],
    'cust_id': [101, 102, 103, 101, 104],
    'order_date': pd.to_datetime(['2019-03-05', '2019-03-15', '2019-02-20', '2019-03-22', '2019-03-29']),
    'order_details': ['Item A', 'Item B', 'Item C', 'Item D', 'Item E'],
    'total_order_cost': [150, 200, 300, 120, 180]
}

# Create DataFrame
df_orders = pd.DataFrame(data)

# Display the DataFrame
print(df_orders)
```

```python
march_orders = df_orders[(df_orders['order_date']>='2019-03-01') & (df_orders['order_date']<='2019-03-30')]
```

```python
march_orders
```

```python
march_orders_total = march_orders[['cust_id','total_order_cost']].groupby(['cust_id']).sum().reset_index()
```

```python
march_orders_total.sort_values(['total_order_cost'], ascending=False)
```

### Solution 2:

```python
MARCH = 3

df_orders['order_year'] = df_orders['order_date'].dt.year
df_orders['order_month'] = df_orders['order_date'].dt.month
```

```python
df_orders
```

```python
march_orders_2 = df_orders[(df_orders['order_year']==2019) & (df_orders['order_month']==MARCH)]
march_orders_2
```

```python
march_total = march_orders_2[['cust_id', 'total_order_cost']].groupby('cust_id', as_index=False).sum().reset_index(drop=True)
```

```python
march_total
```

```python
march_total.sort_values('total_order_cost',ascending=False)
```
