```python
# Plot broadcast indexing
'''
Broadcast indexing simplifies operations on arrays with different shapes and is widely used in numerical computing and data visualization. 
It allows for efficient and concise code when manipulating and plotting data, enabling operations like conditional formatting, grid-based function evaluations, and more.

Suppose you want to highlight all days where the 'High' value was greater than 800 in a different color. Here's how broadcasting can help:

1. Create a condition mask using broadcasting.
2. Use the mask to plot with different styles or colors.
'''

# Assuming df_tesla is already defined and contains the necessary data
# For demonstration, let's create a sample DataFrame similar to df_tesla
dates = pd.date_range(start='2023-01-01', end='2024-06-10', freq='B')
high_values = np.random.uniform(500, 900, len(dates))
high_df = pd.DataFrame({'High': high_values}, index=dates)

# Define a threshold for highlighting
threshold = 800

# Create a mask for the condition
mask = high_df['High'] > threshold

# Plot the main series
plt.figure(figsize=(12, 4))
high_df['High'].plot(xlim=['2023-01-01', '2024-06-10'], ylim=[0, 1000], ls='--', c='green', label='High')

# Plot the highlighted points
plt.scatter(high_df.index[mask], high_df['High'][mask], color='red', label=f'High > {threshold}')

# Add labels and legend
plt.xlabel('Date')
plt.ylabel('High Value')
plt.title('Tesla High Prices with Highlighted Points')
plt.legend()

plt.show()
```

In time series forecasting, rolling and expanding windows are two techniques used to create smaller time series datasets from a larger one. The main difference between the two lies in how they handle the window size and the way they extract features.

### Rolling Window

A rolling window is a fixed-size window that slides over the time series data, extracting features at each step. The window size is constant, and the window moves one time step at a time. This means that the features extracted at each step are based on the same number of previous time steps.

For example, if you have a time series with 100 observations and a rolling window of size 10, the first window would be from observation 1 to 10, the second window would be from observation 2 to 11, and so on.

### Expanding Window

An expanding window, on the other hand, is a window that grows in size as it moves over the time series data. The window size increases by one time step at each step, allowing the model to incorporate more recent data.

For example, if you have a time series with 100 observations and an expanding window, the first window would be from observation 1 to 10, the second window would be from observation 1 to 11, the third window would be from observation 1 to 12, and so on.

### Key differences

Here are the key differences between rolling and expanding windows:

* Window size: Rolling window has a fixed size, while expanding window grows in size as it moves over the data.
* Feature extraction: Rolling window extracts features based on a fixed number of previous time steps, while expanding window extracts features based on a growing number of previous time steps.
* Data incorporation: Rolling window incorporates the same amount of data at each step, while expanding window incorporates more recent data as it moves forward.
When to use each

### Rolling windows are suitable when:

* The time series has a stable trend and seasonality.
* The most recent data is not as important for forecasting.
* You want to extract features based on a fixed number of previous time steps.

### Expanding windows are suitable when:

* The time series is volatile or has a strong trend.
* The most recent data is important for forecasting.
* You want to incorporate more recent data into your model.

In summary, rolling windows are useful when you want to extract features based on a fixed number of previous time steps, while expanding windows are useful when you want to incorporate more recent data into your model.
